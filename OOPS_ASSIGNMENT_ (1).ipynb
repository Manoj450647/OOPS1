### 1. What is Object-Oriented Programming (OOP)?

**Answer:** Object-Oriented Programming is a programming paradigm that uses objects and classes to structure code. It focuses on using reusable code and creating programs based on real-world entities. Key principles of OOP include encapsulation, inheritance, abstraction, and polymorphism.

---

### 2. What is a class in OOP?

**Answer:** A class is a blueprint for creating objects. It defines attributes (variables) and methods (functions) that the created objects (instances) will have.

```python
class Car:
    def __init__(self, brand):
        self.brand = brand
    def drive(self):
        print(f"Driving a {self.brand} car")
```

---

### 3. What is an object in OOP?

**Answer:** An object is an instance of a class. It represents a specific entity with attributes and behaviors defined by the class.

```python
my_car = Car("Toyota")
my_car.drive()
```

---

### 4. What is the difference between abstraction and encapsulation?

**Answer:**

* **Abstraction** hides complex implementation details and shows only the essential features.
* **Encapsulation** hides internal data and enforces access through methods to protect the state of an object.

---

### 5. What are dunder methods in Python?

**Answer:** Dunder (double underscore) methods are special methods with names like `__init__`, `__str__`, etc., used to perform special operations in classes.

```python
def __str__(self):
    return "String representation"
```

### 6. Explain the concept of inheritance in OOP.

**Answer:** Inheritance allows one class to acquire properties and methods of another class. It promotes code reuse.

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Bark")
```

### 7. What is polymorphism in OOP?

**Answer:** Polymorphism allows objects of different classes to be treated as objects of a common superclass. It means the same method can behave differently in different classes.

```python
class Bird:
    def fly(self):
        print("Bird flying")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flying")
```

### 8. How is encapsulation achieved in Python?

**Answer:** Encapsulation is achieved by making attributes private using `__` and accessing them using getter and setter methods.

```python
class BankAccount:

        self.__balance = 0
    def deposit(self, amount):
        self.__balance += amount
    def get_balance(self):
        return self.__balance
```

### 9. What is a constructor in Python?

### 10. What are class and static methods in Python?

**Answer:**

### 11. What is method overloading in Python?

**Answer:** Python doesn't support method overloading directly. Instead, default arguments or variable arguments are used to achieve similar behavior.

### 12. What is method overriding in OOP?

**Answer:** Method overriding occurs when a subclass provides a specific implementation of a method that is already defined in its parent class.

---

### 13. What is a property decorator in Python?

**Answer:** `@property` is used to make a method act like an attribute.

```python
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius
```

### 14. Why is polymorphism important in OOP?

### 15. What is an abstract class in Python?

```python
from abc import ABC, abstractmethod
class Shape(ABC):

        pass
```

````python


## 1. Create a parent class Animal with a method `speak()` that prints a generic message. Create a child class Dog that overrides the `speak()` method to print "Bark!".

class Animal:
    def speak(self):
        print("Animal speaks in a general way")

class Dog(Animal):
    def speak(self):
        print("Bark!")

# Usage
d = Dog()
d.speak()  # Output: Bark!


## 2. Abstract class Shape with method `area()`, derived classes Circle and Rectangle.

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Usage
c = Circle(5)
r = Rectangle(4, 6)
print(c.area())  # Circle Area
print(r.area())  # Rectangle Area

## 3. Multi-level inheritance: Vehicle → Car → ElectricCar

class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery

# Usage
e_car = ElectricCar("Electric", "Tesla", "100 kWh")
print(e_car.type, e_car.brand, e_car.battery)

## 4. Polymorphism with Bird, Sparrow, and Penguin

class Bird:
    def fly(self):
        print("Bird can fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguins can't fly")

# Usage
for bird in (Sparrow(), Penguin()):
    bird.fly()


## 5. Encapsulation with BankAccount

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds")

    def get_balance(self):
        return self.__balance

# Usage
acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.get_balance())


## 6. Runtime polymorphism with Instrument

class Instrument:
    def play(self):
        print("Playing instrument")

class Guitar(Instrument):
    def play(self):
        print("Playing guitar")

class Piano(Instrument):
    def play(self):
        print("Playing piano")

# Usage
for i in (Guitar(), Piano()):
    i.play()


## 7. Class and static methods in MathOperations

class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Usage
print(MathOperations.add_numbers(5, 3))
print(MathOperations.subtract_numbers(10, 4))


## 8. Count total number of persons created

class Person:
    count = 0

    def __init__(self):
        Person.count += 1

    @classmethod
    def get_count(cls):
        return cls.count

# Usage
p1 = Person()
p2 = Person()
print(Person.get_count())


## 9. Override `__str__` in Fraction

class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"

# Usage
f = Fraction(3, 4)
print(f)


## 10. Operator overloading in Vector

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

# Usage
v1 = Vector(2, 3)
v2 = Vector(1, 4)
print(v1 + v2)

## 11. Person class with greet method

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Usage
p = Person("Alice", 25)
p.greet()


## 12. Student average_grade method

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

# Usage
s = Student("Bob", [80, 90, 85])
print(s.average_grade())


## 13. Rectangle with set_dimensions and area

class Rectangle:
    def set_dimensions(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Usage
r = Rectangle()
r.set_dimensions(5, 10)
print(r.area())


## 14. Employee and Manager with salary and bonus

class Employee:
    def __init__(self, hours, rate):
        self.hours = hours
        self.rate = rate

    def calculate_salary(self):
        return self.hours * self.rate

class Manager(Employee):
    def __init__(self, hours, rate, bonus):
        super().__init__(hours, rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

# Usage
m = Manager(40, 50, 500)
print(m.calculate_salary())


## 15. Product class with total_price method

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

# Usage
p = Product("Pen", 10, 5)
print(p.total_price())


## 16. Animal abstract class with sound method

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

# Usage
c = Cow()
s = Sheep()
c.sound()
s.sound()


## 17. Book class with `get_book_info()`

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"{self.title} by {self.author}, published in {self.year_published}"

# Usage
b = Book("Python Basics", "John Doe", 2022)
print(b.get_book_info())


## 18. House and Mansion

class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

# Usage
m = Mansion("123 Street", 5000000, 10)
print(m.address, m.price, m.number_of_rooms)
````
